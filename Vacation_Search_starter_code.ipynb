```python
# 1. Import the WeatherPy_database.csv file. 
city_data_df = pd.read_csv("Weather_Database/WeatherPy_database.csv")
city_data_df.head()
```

```python
# 2. Prompt the user to enter minimum and maximum temperature criteria 

min_temp  = float(input ("Enter minimum temperature:"))
max_temp  = float(input ("Enter maximum temperature:"))

```

```python
# 3. Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.

temp_filter_data_df= city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) | (city_data_df["Max Temp"] >= min_temp)]

temp_filter_data_df.head()
#city_data_df["Max Temp"]
```

```python
# 4a. Determine if there are any empty rows.
temp_filter_data_df.isnull().sum()
```

```python
# 4b. Drop any empty rows and create a new DataFrame that doesnâ€™t have empty rows.
temp_filter_data2_df=temp_filter_data_df.dropna()
temp_filter_data2_df.isnull().sum()
```

```python
# 5a. Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = temp_filter_data2_df[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()


# 5b. Create a new column "Hotel Name"
hotel_df["Hotel Name"] = ""
hotel_df.tail(10)
```

```python
# Set the parameters to search for a hotel in Paris.
params = {
    "radius": 10000,
    "types": "lodging",
    "key": g_key,
    "location": "42.6001, -73.9662"}
# Use base URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
# Make request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()

hotels
```
